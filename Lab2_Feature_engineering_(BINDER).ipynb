```python
df=pd.read_csv('healthcare_data-2(BINDER).csv')

df
```

```python
from sklearn.preprocessing import LabelEncoder

#Creating a instance of label Encoder
le = LabelEncoder()

#Using .fit_transform function to fit label
#encoder and return encoded label
label = le.fit_transform(df['Outcome'])

#printing label
label
```

```python
#Remove the column 'Outcome' from df
#as it is of no use now.
df.drop("Outcome", axis=1, inplace=True)

#Appending the array to our dataFrame
#with column name 'Outcome'
df["Outcome"] = label

#print dataframe
df
```

```python
old_cols=df.columns.values
new_cols=['Pregnancies','BloodPressure','Glucose','Insulin','SkinThickness','BMI','DiabetesPedigreeFunction','Age','Outcome']
df=df.reindex(columns=new_cols)
df
```
